```python
import requests
import holoviews as hv
import pandas as pd
import numpy as np
pd.options.plotting.backend = 'holoviews'
%matplotlib inline
```

```python
def get_crypto(start_date="1/1/2020",
    end_date="3/1/2020",
    crypto="BTC-USD"):
    rest_api_str = f'https://api.pro.coinbase.com/products/{crypto}/candles?start="{start_date}"&end="{end_date}"&granularity=86400'
    response = requests.get(rest_api_str)
    crypto_df = pd.DataFrame(response.json())
    crypto_df=crypto_df.iloc[:,[0,4,5]]
    crypto_df.columns=['date','close','volume']
    crypto_df["date"] = pd.to_datetime(crypto_df["date"], unit='s')
    
    
    return crypto_df
    
#rest_api_str = f'https://api.pro.coinbase.com/products/{crypto}/candles?start="{start_date}"&end="{end_date}"&granularity=86400'
```
